```python
# filename = 'corpus/10600.txt'
def get_tokenize_pos(in_filename,out_filename):
#     filename = f'corpus/filename'
#     print(out_filename)
    in_filename = os.path.abspath(in_filename)
    out_filename = os.path.abspath(out_filename)
    lines = []
    with codecs.open(in_filename, "r",'utf8') as infile:
        lines = infile.readlines()
    turn = 1
    dialogues = []
    for line in lines:
        line = line.strip()
#         print(line)
        speaker = line[:1]
        open_paren_index = line.index('(')
        close_paren_index = line.index(')')
        act = line[open_paren_index+1:close_paren_index]
        text = nltk.word_tokenize(line[line.index(':')+2:].strip())
        tags = nltk.pos_tag(text)
        dialogue = {
            'speaker': speaker,
            'turn': turn,
            'act': act,
            'text': text,
            'tags': tags
        }
#         print(dialogue)
        dialogues.append(dialogue)
        turn += 1
#     outfile = open(out_filename,'wb')
    with open(out_filename, 'wb') as outfile:
        pickle.dump(dialogues,outfile)
#     outfile.close()
    return dialogues
```

```python
mypath = 'corpus'
filenames = [f for f in listdir(mypath) if isfile(join(mypath, f))]

for filename in filenames:
    print(filename)
    out_filename = f'pos_token/{filename[:-4]}.pickle'
    get_tokenize_pos(f'corpus/{filename}',out_filename)
```
